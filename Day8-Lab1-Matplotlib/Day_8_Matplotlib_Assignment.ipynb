```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
plt.plot([1, 2, 3])
plt.plot(pd.DataFrame([4, 5, 6]));
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie([1, 2, 3], explode=[0.1, 0.1, 0.1]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter([1, 2, 3, 8, 9, 9], [20, 3, 80, 77, 76, 30], marker="o", alpha=0.4, s=50);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.title("plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.scatter([1, 2, 3, 8, 9, 9], [20, 3, 80, 77, 76, 30], marker="o", alpha=0.4, s=50);
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1, 2, 3], c="b")
plt.plot([30, 32, 9, 12], c="c")
plt.plot(pd.DataFrame([4, 5, 6]), c="m")
plt.legend(["label1", "label2", "label3"], loc="upper left");
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], c="b")
plt.subplot(2, 2, 2)
plt.plot([30, 32, 9, 12], c="c")
plt.subplot(2, 2, 3)
plt.plot(pd.DataFrame([4, 5, 6]), c="m");
```
